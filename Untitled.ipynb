```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import r2_score, mean_absolute_error
import pickle
```

```python
# 1. Load the generated dataset
df = pd.read_csv("UP_Property_Dataset.csv")

```

```python
# 2. Encoding Categorical Data
# Hum har category (City, Area, etc.) ke liye alag encoder save karenge
encoders = {}
categorical_cols = ['City', 'Area', 'Property_Type', 'Facing']

for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    encoders[col] = le
```

```python

# 3. Features aur Target set karein
x = df.drop('Price', axis=1)
y = df['Price']

```

```python
# 4. Train-Test Split (90% training, 10% testing)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)

```

```python
# 5. XGBoost Model Training
# Max depth 10 rakha hai taaki Area-wise patterns achhe se seekh sake
model = xgb.XGBRegressor(
    n_estimators=1000,
    max_depth=10,
    learning_rate=0.05,
    verbosity=1,
    objective='reg:squarederror'
)

print("🚀 Training started... thoda sabr rakho...")
model.fit(X_train, y_train)
```

```python
# 6. Accuracy Check
y_pred = model.predict(X_test)
accuracy = r2_score(y_test, y_pred) * 100
mae = mean_absolute_error(y_test, y_pred)

print(f"\n✅ Training Complete!")
print(f"📊 Accuracy: {accuracy:.2f}%")
print(f"💸 Mean Error: ₹{int(mae)}")

```
